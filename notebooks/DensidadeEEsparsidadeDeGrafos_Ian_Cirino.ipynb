# Densidade e Esparsidade de grafos
## Conceito de densidade de um grafo:
A densidade $ϱ$ de um grafo simples é dada pela fração entre o número de arestas de um grafo e o número de arestas máximas possiveis nesse grafo como mostrada pela equação (*).

$ϱ = \frac{m}{m_máx} = \frac{2m}{n(n-1)}$ (*)

O máximo número de arestas em um grafo simples pode ser dado pela equação (**).

$m_máx = \binom{n}{2} = \frac{n!}{(n-2)!2!} = \frac{1}{2} n(n-1) $ (**)

Como o número de arestas sempre é menor ou igual ao número de arestas totais ($m \leq m_máx$) então a densidade sempre vai estar no intervalo entre 0 e 1 ($0 \leq ϱ \leq 1$).
Um resultado interessante é que: quando temos um grafo simples de vertices de mesmo grau $c$ a formula (*) pode ser reduzida para $ ϱ = \frac{c}{n-1}$ sendo $c = \frac{2m}{n}$. Isso é provado com mais detalhes na seção 6.9 do livro Networks, Newman.

### Exemplo de algoritmos de densidade:



```python
# Grafo
grafo = {1: [3, 2], 2: [1, 4], 3: [1, 4], 4: [3, 2, 5], 5: [4]}

# Contador de vértices e arestas
n = len(grafo)
m = 0
for i in grafo:
    m += len(grafo[i])

# Como cada aresta é contada duas vezes, dividimos por 2
m = m // 2

print(f"n = {n}")
print(f"m = {m}")

# Cálculo da densidade
m_max = (1 / 2) * n * (n - 1)
if n == 0:
    print("grafo sem vertices")
else:
    p = m / m_max

print(f"Densidade p = {p}")
```

## Grafos esparsos e densos:
Uma rede onde ρ → 0 no limite
de grande n é dita esparsa( $ n \implies ∞ $ ; $ϱ \implies 0$). E dizemos que ele é denso se a fração de elementos diferentes de zero na matriz de adjacência é não nula no limite de grande n.
Para um grafo esparso sua matriz de adjacência A se aproxima da matriz nula. Isso pode ser visto no exemplo abaixo.

###Exemplo da relação entre matriz de adjacência e a densidade de um grafo:


```python
# Exemplo de matriz de adjacência esparsa
Adj = [[1 if i == j + 1 else 0 for j in range(60)] for i in range(60)]

n = len(Adj)

# Calcula o número de arestas
m = 0
for i in range(n):
    for j in range(n):
        m += Adj[i][j]

# Calcula a densidade
m_max = (1/2) * n * (n - 1)
p = m / m_max

# Exibe a matriz e a densidade
for i in range(n):
    print(Adj[i])
print(f"Densidade: {p}")
```
