# 1. O que é graph-tool?

O *graph-tool* é uma biblioteca em Python feita para trabalhar com grafos de forma muito eficiente.

Ela é **útil quando você precisa manipular grandes redes de dados e realizar análises estatísticas sobre elas**, como detectar comunidades dentro da rede, calcular métricas, entre outras coisas.

O grande diferencial é que essa biblioteca **é muito rápida porque usa C++ "por trás" do Python para processar tudo**.

# 2. Como o graph-tool pode me ajudar?

**1. Detecção de Comunidades:** Ele facilita encontrar grupos conectados dentro de uma rede (como grupos de amigos em redes sociais) usando métodos avançados que evitam erros comuns de análise.

**2. Processamento Paralelo:** Aproveita o poder de múltiplos núcleos de processadores para acelerar a execução de algoritmos em redes grandes, tornando o processamento muito mais rápido.

**3. Visualização de Redes:** Permite criar visualizações interativas e estáticas das suas redes, além de animações, ajudando a entender melhor as conexões e a estrutura dos dados.

**4. Filtragem de Grafos:** Você pode ocultar temporariamente partes da rede sem perder desempenho, facilitando o foco em áreas específicas durante a análise.

**5. Integração com Repositórios de Dados:** Oferece integração com repositórios de dados prontos (Netzschleuder), facilitando o carregamento de redes sem precisar preparar tudo do zero.

```python
g = Graph() # Cria um grafo vazio
```

* **Grafo direcionado (ou dígrafo):** as arestas têm uma direção, ou seja, conectam um vértice a outro em uma ordem específica. Exemplo: grafos que representam links na web ou fluxos de trânsito.

* **Grafo não direcionado:** as arestas não têm direção, ou seja, a conexão entre os vértices é bidirecional. Exemplo: grafos que representam amizades ou conexões em redes de comunicação.

Por padrão, os grafos criados são direcionados. Para criar grafos não direcionados, você precisa passar o parâmetro `directed=False`:

```python
ug = Graph(directed=False) # Cria um grafo não direcionado
```

Você pode alternar entre grafos direcionados e não direcionados a qualquer momento usando o método `set_directed()`:

```python
ug = Graph()
ug.set_directed(False)  # Torna o grafo não direcionado
assert ug.is_directed() == False  # Verifica se o grafo é não direcionado
```

### 6.7 Adicionando Atributos a Nós e Arestas com `Property Maps`

```python
from graph_tool.all import *

g = Graph()  # Cria um grafo vazio

# Adiciona dois vértices
v1 = g.add_vertex()
v2 = g.add_vertex()

# Adiciona uma aresta entre os dois vértices
e = g.add_edge(v1, v2)
```

Agora, criaremos `Property Maps` para adicionar um peso às arestas e uma cor aos vértices:

* `eweight`: Um `EdgePropertyMap` para armazenar o peso das arestas (números decimais).
* `vcolor`: Um `VertexPropertyMap` para armazenar a cor dos vértices (strings que representam cores).

```python
# Cria Property Maps
eweight = g.new_ep("double")    # Para armazenar pesos de arestas
vcolor = g.new_vp("string")     # Para armazenar cores dos vértices
```

Agora, atribuímos os valores de peso e cor para as arestas e vértices:

```python
# Define o peso da aresta
eweight[e] = 25.3

# Define as cores dos vértices
vcolor[v1] = "#1c71d8"  # Azul
vcolor[v2] = "#2ec27e"  # Verde
```

Utilize as `Property Maps` para personalizar a visualização do grafo:

```python
graph_draw(g, vertex_text=g.vertex_index,
           vertex_fill_color=vcolor,    # Define as cores dos vértices
           edge_pen_width=eweight,      # Define a espessura das arestas
           output="two-nodes-color.png") # Salva o grafo como SVG
```

#### 6.8.1 Criando um Grafo com Vértices e Arestas Usando Lista de Arestas

No `graph-tool`, é possível adicionar muitos vértices e arestas ao mesmo tempo de maneira eficiente. Você pode criar um grafo diretamente a partir de uma lista de arestas:

```python
g = Graph([('foo', 'bar'), ('gnu', 'gnat')], hashed=True)
```

* `hashed=True`: Permite usar valores de vértices como strings ou outros objetos ao invés de apenas inteiros.
* Nesse exemplo, os vértices `'foo'`, `'bar`', `'gnu`' e `'gnat'` serão criados com arestas entre eles.


#### 6.8.2 Adicionando Propriedades às Arestas ao Criar o Grafo

Podemos criar o grafo e adicionar propriedades como peso e tempo às arestas:

```python
g = Graph([('foo', 'bar', .5, 1), ('gnu', 'gnat', .78, 2)], hashed=True,
          eprops=[('weight', 'double'), ('time', 'int')])
```

* `eprops`: Lista de propriedades das arestas. Aqui, estamos adicionando propriedades de peso (weight) e tempo (time).
* Os valores `.5`,` 1`, `.78`, e `2` correspondem aos valores dessas propriedades para as respectivas arestas.

#### 6.8.3 Criando um Grafo Usando Lista de Adjacências

Você também pode criar um grafo a partir de uma lista de adjacências, que é um dicionário que mapeia os vizinhos de cada vértice:

```python
g = Graph({0: [2, 3], 1: [4], 3: [4, 5], 6: []})
```

Neste exemplo, o vértice 0 tem arestas para 2 e 3, o vértice 1 para 4, e assim por diante.

#### 6.8.4 Criando um Grafo Usando Matriz de Adjacência

Para criar um grafo usando uma matriz de adjacência, você pode convertê-la em uma matriz esparsa do `scipy`:

$$
\begin{bmatrix}
0 & 1.2 & 0 \\
0 & 0 & 10 \\
0 & 7 & 0
\end{bmatrix}
$$

Essa é a matriz de adjacência 3×3, onde os elementos diferentes de zero representam conexões (ou pesos de arestas) entre vértices. O valor 0 significa que não há aresta entre os dois vértices.


```python
import numpy as np
import scipy.sparse # Esse submódulo oferece suporte para matrizes esparsas em Python

# Cria uma matriz de adjacência
m = np.array([[0, 1.2, 0],
              [0, 0, 10],
              [0, 7, 0]])

# Converte para uma matriz esparsa e cria o grafo
g = Graph(scipy.sparse.lil_matrix(m))

# Imprime os pesos das arestas
print(g.ep.weight.a)
```

Nesse caso, a matriz define as conexões entre vértices, e os valores não-zero representam o peso das arestas.

Uma **matriz esparsa** é uma matriz onde **a maioria dos elementos são zeros**. **Ao invés de armazenar todos os elementos, inclusive os zeros, uma matriz esparsa só armazena os valores não nulos e suas localizações**. Usar uma matriz esparsa para representar apenas os valores diferentes de zero é **muito mais eficiente**.

#### 6.8.5 Adicionando Arestas Após Criar o Grafo

Você pode adicionar várias arestas de uma vez usando o método `add_edge_list()`:

```python
g.add_edge_list([(0, 1), (2, 3)])
```

Aqui, arestas são adicionadas entre os pares de vértices (0, 1) e (2, 3).

#### 6.8.6 Usando Vértices Não Inteiros com `hashed=True`

Caso queira usar valores de vértices que não são inteiros, como strings ou objetos Python, você pode fazer isso ao passar `hashed=True`:

```python
g.add_edge_list([('foo', 'bar'), ('gnu', 'gnat')], hashed=True, hash_type="string")
```

#### 6.8.7 Exemplo Prático

Com essas técnicas, você pode manipular grafos grandes e complexos de maneira eficiente no `graph-tool`.

```python
from graph_tool.all import *
from IPython.display import SVG, display

# Cria o grafo a partir de uma lista de arestas com propriedades de peso e tempo
g = Graph([('A', 'B', 0.5, 10), ('B', 'C', 0.8, 5), ('C', 'A', 1.2, 3)],
          hashed=True,
          eprops=[('weight', 'double'), ('time', 'int')])

# Adiciona propriedades de cores para os vértices
vcolor = g.new_vp("string")

# Define cores específicas para alguns vértices
vcolor[g.vertex(0)] = "#ff0000"  # Vértice 'A' em vermelho
vcolor[g.vertex(1)] = "#00ff00"  # Vértice 'B' em verde
vcolor[g.vertex(2)] = "#0000ff"  # Vértice 'C' em azul

# Atribui uma cor padrão para todos os outros vértices
for v in g.vertices():
    if vcolor[v] == "":
        vcolor[v] = "#cccccc"  # Cor cinza para vértices sem cor definida

# Visualiza o grafo em um SVG, personalizando cores e espessura das arestas
graph_draw(g,
           vertex_text=g.vertex_index,  # Mostra o índice do vértice
           vertex_fill_color=vcolor,    # Usa o mapa de cores dos vértices
           edge_pen_width=g.ep.weight,  # Espessura da aresta de acordo com o peso
           output_size=(300, 300),      # Define o tamanho da imagem
           output="graph-example.svg")

# Exibe o arquivo SVG
display(SVG('graph-example.svg'))
```

#### 6.8.8 Removendo Vértices e Arestas

Você pode remover um vértice ou uma aresta usando os métodos `remove_vertex()` e `remove_edge()`:

```python
from graph_tool.all import *

# Cria um grafo
g = Graph()

# Adiciona 5 vértices
v1 = g.add_vertex()
v2 = g.add_vertex()
v3 = g.add_vertex()
v4 = g.add_vertex()
v5 = g.add_vertex()

# Adiciona arestas entre os vértices
e1 = g.add_edge(v1, v2)
e2 = g.add_edge(v2, v3)
e3 = g.add_edge(v3, v4)
e4 = g.add_edge(v4, v5)

print("Grafo antes das remoções:")
print(f"Número de vértices: {g.num_vertices()}")
print(f"Número de arestas: {g.num_edges()}")

# Remove uma aresta (e2: entre v2 e v3)
g.remove_edge(e2)

print("\nApós remover a aresta entre v2 e v3:")
print(f"Número de vértices: {g.num_vertices()}")
print(f"Número de arestas: {g.num_edges()}")

# Remove um vértice (v2)
g.remove_vertex(v2)

print("\nApós remover o vértice v2:")
print(f"Número de vértices: {g.num_vertices()}")
print(f"Número de arestas: {g.num_edges()}")
```

**Nota:** Remover um vértice pode reordenar os índices dos outros vértices.

#### 6.8.10 Filtrando Grafos

Você pode filtrar um grafo para trabalhar com uma visão reduzida. Exemplo, filtrando arestas de alta centralidade de intermediação:

```python
from graph_tool.all import *
from IPython.display import SVG, display

# Cria o grafo e adicionar vértices e arestas
g = Graph()
v1 = g.add_vertex()  # Vértice 0
v2 = g.add_vertex()  # Vértice 1
v3 = g.add_vertex()  # Vértice 2
v4 = g.add_vertex()  # Vértice 3

e1 = g.add_edge(v1, v2)  # Aresta de 0 para 1
e2 = g.add_edge(v2, v3)  # Aresta de 1 para 2
e3 = g.add_edge(v3, v4)  # Aresta de 2 para 3
e4 = g.add_edge(v1, v4)  # Aresta de 0 para 3

# Calcula a centralidade de intermediação
vb, eb = betweenness(g)

# Filtra arestas com baixa intermediação
u = GraphView(g, efilt=eb.fa < 1e-6)

# Gera um layout automático para os vértices
pos = sfdp_layout(u)  # Calcula o layout dos vértices

# Desenha o grafo filtrado e redimensionar
graph_draw(u, pos=pos, output_size=(300, 300), output="filtered_graph.svg")  # Redimensiona para 800x800 pixels

# Exibe o SVG
display(SVG('filtered_graph.svg'))
```

```python
from graph_tool.all import *
from IPython.display import SVG, display

# Cria o grafo e adiciona vértices e arestas
g = Graph()
v1 = g.add_vertex()  # Vértice 0
v2 = g.add_vertex()  # Vértice 1
v3 = g.add_vertex()  # Vértice 2
v4 = g.add_vertex()  # Vértice 3

e1 = g.add_edge(v1, v2)  # Aresta de 0 para 1
e2 = g.add_edge(v2, v3)  # Aresta de 1 para 2
e3 = g.add_edge(v3, v4)  # Aresta de 2 para 3
e4 = g.add_edge(v1, v4)  # Aresta de 0 para 3

# Adiciona propriedades de cor aos vértices e peso às arestas
vcolor = g.new_vertex_property("string")
vcolor[v1] = "#ff0000"  # Vértice 0 vermelho
vcolor[v2] = "#00ff00"  # Vértice 1 verde
vcolor[v3] = "#0000ff"  # Vértice 2 azul
vcolor[v4] = "#ffff00"  # Vértice 3 amarelo

eweight = g.new_edge_property("double")
eweight[e1] = 2.5
eweight[e2] = 1.0
eweight[e3] = 3.0
eweight[e4] = 4.5

# Filtra o grafo para remover arestas com peso maior que 2.5
edge_filter = g.new_edge_property("bool")
for e in g.edges():
    edge_filter[e] = eweight[e] <= 2.5
g.set_edge_filter(edge_filter)

# Remove o vértice 3 (índice 2)
g.remove_vertex(g.vertex(2))

# Visualiza o grafo
graph_draw(g, vertex_text=g.vertex_index, vertex_fill_color=vcolor,
           edge_pen_width=eweight, output_size=(300, 300),
           output="graph_filtered.svg")

# Exibe o SVG
display(SVG('graph_filtered.svg'))
```

# 7. Netzschleuder  // "nêts-shloy-der"

* Pode ser usado para análises de redes sociais, biológicas, tecnológicas, entre outras.

### 7.2 Como usar o Netzschleuder no `graph-tool`

#### 7.2.1 Importação

Primeiro, você precisa importar o graph-tool e o módulo específico que lida com a integração ao repositório Netzschleuder.

```python
from graph_tool.all import *
from graph_tool.collection import ns # O módulo ns fornece acesso ao repositório Netzschleuder.
```

#### 7.2.2 Carregar uma Rede do Netzschleuder

Para carregar uma rede específica do repositório, basta acessar pelo nome da rede dentro do módulo `ns`. Por exemplo, podemos carregar a rede de interações entre personagens de "Os Miseráveis".

```python
# Carrega a rede "lesmis" do repositório Netzschleuder
g = ns["lesmis"]
```

Essa rede contém os personagens como nós (vértices) e as interações entre eles como arestas.

#### 7.2.3 Explorar a Estrutura da Rede

Antes de manipularmos a rede, podemos inspecionar algumas características, como o número de vértices (personagens) e arestas (interações).

```python
# Número de vértices e arestas
print(f"Número de personagens: {g.num_vertices()}")
print(f"Número de interações: {g.num_edges()}")
```

#### 7.2.4 Visualizar a Rede

Agora, vamos visualizar essa rede simples. Para isso, precisamos calcular um layout de visualização para os vértices usando a função `sfdp_layout(`).

```python
from graph_tool.all import *
from IPython.display import SVG, display

# Calcula o layout dos vértices
pos = sfdp_layout(g)

# Desenha o grafo com o layout calculado e salvar como arquivo SVG
graph_draw(g, pos=pos, output_size=(500, 500), output="lesmis_network.svg")

# Exibe o arquivo SVG
display(SVG('lesmis_network.svg'))
```

#### 7.2.5 Calcular Propriedades da Rede

Agora podemos calcular algumas propriedades da rede, como o grau de cada vértice (quantas interações cada personagem tem) e a centralidade de intermediação (betweenness centrality).

* **Calcular Graus dos Vértices:** O grau de um vértice indica quantas interações um personagem tem com outros personagens.

```python
# Imprime o grau de cada vértice (personagem)
for v in g.vertices():
    print(f"Personagem {int(v)} tem grau {v.out_degree()}")
```

* **Calcular Centralidade de Intermediação:** Vamos calcular a centralidade de intermediação, que mede o quão importante um vértice é em termos de conectar diferentes partes da rede. Em outros termos, **mede o quanto um personagem atua como um intermediário nas interações entre outros personagens**.

## 8.2 Criando o Grafo

Rede de Transporte Aéreo

```python
import csv
from graph_tool.all import *

# 1. Criar um grafo vazio
graph = Graph(directed=True)

# Criar propriedades para armazenar os nomes dos aeroportos
airport_name = graph.new_vertex_property("string")
airport_id_map = {}

# 2. Carregar os dados do arquivo `routes.dat`
with open("routes.dat", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        # Os índices do arquivo CSV, conforme o formato de dados do OpenFlights
        source_airport_id = row[3]
        destination_airport_id = row[5]

        # Verificar se os IDs dos aeroportos são válidos (diferentes de "\\N")
        if source_airport_id == "\\N" or destination_airport_id == "\\N":
            continue

        # Adicionar vértices ao grafo, se necessário
        if source_airport_id not in airport_id_map:
            v_source = graph.add_vertex()
            airport_id_map[source_airport_id] = v_source
        else:
            v_source = airport_id_map[source_airport_id]

        if destination_airport_id not in airport_id_map:
            v_dest = graph.add_vertex()
            airport_id_map[destination_airport_id] = v_dest
        else:
            v_dest = airport_id_map[destination_airport_id]

        # Adicionar uma aresta entre o aeroporto de origem e destino
        graph.add_edge(v_source, v_dest)

# 3. Salvar a rede
graph.save("rede_aerea.gt")
print("Rede de transporte aéreo carregada e salva com sucesso.")
```

### Robustez e Vulnerabilidade da Rede


* **O que é robustez?** Capacidade de continuar funcionando mesmo quando alguns de seus elementos (nós ou conexões) falham.
* **Por que analisar?** Crucial para prever **como o sistema reagiria ao fechamento de um aeroporto importante** (por exemplo, devido a uma emergência ou desastre natural).
* **Como fazer?** Simular a remoção dos principais hubs e analisar como o tamanho do maior componente conectado muda. Realizamos a **análise de assortatividade para entender como aeroportos de diferentes tamanhos se conectam**.

> Os nós tendem a se conectar a outros nós com graus semelhantes, ou seja, características semelhantes. Esta característica é conhecida como **assortatividade**.

```python
from graph_tool.all import *
import numpy as np
import matplotlib.pyplot as plt

# Carregar o grafo salvo
graph = load_graph("rede_aerea.gt")

# Calcular o grau total (entrada + saída) para cada vértice
total_degree = graph.degree_property_map("total")

# Identificar os 10 principais hubs com maior grau total
top_hubs = sorted(graph.vertices(), key=lambda v: total_degree[v], reverse=True)[:10]

# Lista para armazenar o tamanho do maior componente após cada remoção
component_sizes = []

# Função para calcular o tamanho do maior componente conectado
def calculate_largest_component_size(g):
    largest_component = label_largest_component(g)
    return np.sum(largest_component.a)

# Salvar o tamanho do maior componente antes de remover qualquer vértice
initial_size = calculate_largest_component_size(graph)
component_sizes.append(initial_size)

# Simular a remoção dos hubs, um por um
for hub in top_hubs:
    # Remover o hub do grafo
    graph.clear_vertex(hub)  # Remove todas as arestas conectadas ao vértice
    graph.remove_vertex(hub)  # Remove o vértice em si

    # Calcular o tamanho do maior componente conectado após a remoção
    size_after_removal = calculate_largest_component_size(graph)
    component_sizes.append(size_after_removal)

# Exibir os resultados da simulação
print(f"Tamanho inicial do maior componente: {initial_size}")
for i, size in enumerate(component_sizes[1:], start=1):
    print(f"Tamanho do maior componente após remover {i} hub(s): {size}")

# Visualizar a mudança no tamanho do maior componente
plt.figure(figsize=(8, 5))
plt.plot(range(len(component_sizes)), component_sizes, marker='o', linestyle='-', color='b')
plt.xlabel("Número de Hubs Removidos")
plt.ylabel("Tamanho do Maior Componente Conectado")
plt.title("Impacto da Remoção dos Principais Hubs na Conectividade da Rede")
plt.grid(True)
plt.show()

# Calcular a assortatividade por grau
degree_assortativity = assortativity(graph, "total")
print(f"Coeficiente de assortatividade por grau: {degree_assortativity}")
```

* À medida que removemos os principais aeroportos (hubs mais conectados), o tamanho do maior grupo interligado diminui, mas de forma lenta.
* Após remover 10 dos principais hubs, ainda restam 3226 aeroportos conectados no maior grupo.
* **O que isso mostra?** A rede é resistente e mesmo quando os maiores hubs são removidos, a maioria dos aeroportos ainda permanece conectada.

### Eficiência e Otimização do Tráfego Aéreo


* **O que é eficiência?** Quão bem a rede utiliza suas rotas para conectar os aeroportos, minimizando o número de conexões necessárias entre dois pontos.
* **Por que analisar?** Identificar gargalos e oportunidades de otimização, como a necessidade de novas rotas para melhorar a cobertura.
* **Como fazer?** Podemos analisar coeficiente de clustering (para ver se há muitos triângulos na rede) e eficiência de rede (avaliando a rapidez com que a informação se espalha pela rede).


```python
from graph_tool.all import *
import numpy as np

# Carregar o grafo salvo
graph = load_graph("rede_aerea.gt")

# 1. Calcular o coeficiente de clustering global
clustering_coefficient = global_clustering(graph)
print(f"Coeficiente de clustering global: {clustering_coefficient}")

# 2. Calcular a eficiência da rede
# A eficiência é a média do inverso das distâncias entre todos os pares de vértices
def calculate_efficiency(graph):
    efficiency = 0.0
    num_pairs = 0
    distances = shortest_distance(graph)  # Calcula todas as distâncias mais curtas
    num_vertices = graph.num_vertices()

    for i in range(num_vertices):
        for j in range(i + 1, num_vertices):
            if distances[i][j] != float('inf'):  # Considera apenas pares conectados
                efficiency += 1.0 / distances[i][j]
                num_pairs += 1

    # A eficiência é a soma dos inversos das distâncias, dividida pelo número de pares
    if num_pairs > 0:
        return efficiency / num_pairs
    else:
        return 0

network_efficiency = calculate_efficiency(graph)
print(f"Eficiência da rede: {network_efficiency}")
```

* Um coeficiente de clustering de aproximadamente 0.34 indica que a rede aérea tem uma tendência moderada de formar triângulos, ou seja, **os aeroportos têm vizinhos que também tendem a estar conectados entre si.**

* A eficiência da rede de aproximadamente 0.26 sugere que, em média, **as conexões entre os aeroportos não são diretas, mas ainda relativamente curtas**.

### Análise Espectral da Rede


* **O que é?** A análise espectral usa os autovalores da matriz Laplaciana da rede para entender características estruturais.
* **Por que analisar**? Autovalores baixos podem indicar clusters ou regiões da rede que são fortemente conectadas internamente, mas menos conectadas ao resto da rede.
* **Como fazer?** Calcular os autovalores e autovetores da matriz Laplaciana e interpretar sua distribuição.

> **Matriz Laplaciana de um grafo é dada por L = D - A, onde D é a matriz diagonal de graus e A é a matriz de adjacência.** Usando o graph-tool, a função laplacian() deve gerar uma matriz Laplaciana adequada.



```python
from graph_tool.all import *
import numpy as np
import matplotlib.pyplot as plt

# Carregar o grafo salvo
graph = load_graph("rede_aerea.gt")

# Certificar-se de que o grafo é tratado como não-direcionado para cálculo da Laplaciana
graph.set_directed(False)

# 1. Calcular a matriz Laplaciana da rede
laplacian_matrix = laplacian(graph)

# 2. Calcular os autovalores e autovetores da matriz Laplaciana
eigenvalues, eigenvectors = np.linalg.eigh(laplacian_matrix.toarray())

# 3. Exibir os primeiros 10 autovalores (ordenados)
print("Primeiros 10 autovalores da matriz Laplaciana (corrigida):")
print(eigenvalues[:10])

# 4. Visualizar a distribuição dos autovalores em um gráfico
plt.figure(figsize=(10, 6))
plt.plot(range(len(eigenvalues)), eigenvalues, marker='o', linestyle='-', color='blue', alpha=0.7)
plt.xlabel('Índice do Autovalor')
plt.ylabel('Autovalor')
plt.title('Distribuição dos Autovalores da Matriz Laplaciana (Corrigida)')
plt.grid(True)
plt.show()
```

> O **Fiedler Value (Segundo Menor Autovalor)** mostra o **quão fácil é dividir uma rede em duas partes**. Se ele estiver perto de zero, significa que a rede tem um ponto fraco, como se pudesse ser dividida em dois grupos grandes com poucas conexões entre eles.

> O **Fiedler Vector é como um mapa para essa divisão**. Ele **mostra em qual grupo cada vértice (como um aeroporto) deve ficar**: valores positivos e negativos do vetor indicam os dois grupos.

```python
# 5. Identificar o segundo menor autovalor (Fiedler value) e o vetor associado (Fiedler vector)
fiedler_value = eigenvalues[1]
fiedler_vector = eigenvectors[:, 1]

print(f"Segundo menor autovalor (Fiedler value): {fiedler_value}")

# 6. Usar o Fiedler vector para particionar a rede em duas comunidades
# Vértices com valores positivos e negativos no Fiedler vector
partition = fiedler_vector > 0

# Criar uma propriedade de cor para visualizar os clusters detectados
vertex_color = graph.new_vertex_property("int")
for i, v in enumerate(graph.vertices()):
    vertex_color[v] = int(partition[i])

# Visualizar a divisão da rede com base no Fiedler vector
output_file = "fiedler_partition.png"
pos = sfdp_layout(graph)
graph_draw(
    graph,
    pos=pos,
    vertex_fill_color=vertex_color,
    output_size=(800, 800),
    output=output_file
)

# Exibir a imagem usando matplotlib
img = plt.imread(output_file)
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
plt.title("Divisão da Rede Usando o Fiedler Vector")
plt.show()
```

Indica que **a rede é muito fácil de dividir em duas parte**s, sugerindo a presença de um "corte natural."
